```python
%matplotlib inline

import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
map_GJ= gpd.GeoDataFrame.from_file('./LSMD_CONT_LDREG_서울_광진구/LSMD_CONT_LDREG_11215.shp', encoding='cp949')
map_GJ
```

```python
print(map_GJ.crs)
```
