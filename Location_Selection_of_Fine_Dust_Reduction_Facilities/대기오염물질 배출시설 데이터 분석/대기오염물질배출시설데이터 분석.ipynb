## 기초통계량 분석

```python
result_dataset = dataset_GJ[using_columns]
result_dataset
```

```python
result_dataset.info()
```

```python
result_dataset = result_dataset.copy()
result_dataset['배출시설조업시간']=result_dataset['배출시설조업시간'].astype(float)
result_dataset['좌표정보(x)']=result_dataset['좌표정보(x)'].astype(float)
result_dataset['좌표정보(y)']=result_dataset['좌표정보(y)'].astype(float)
```

```python
result_dataset.describe()
```

```python
result_dataset['종별명'].value_counts()
```

```python
fig = plt.figure(figsize = (16,12))

axes1 = fig.add_subplot(2,2,1)
axes1.hist(result_dataset['배출시설조업시간'], bins=20)
axes1.set_title("Polution Running Hour")
axes1.set_xlabel("running hour")
axes1.set_ylabel("count")

axes2 = fig.add_subplot(2,2,2)
axes2.hist(result_dataset['배출시설연간가동일수'], bins=20)
axes2.set_title("Polution Running Day")
axes2.set_xlabel("running day")
axes2.set_ylabel("count")

axes3 = fig.add_subplot(2,2,3)
axes3.hist(result_dataset['방지시설조업시간'], bins=20)
axes3.set_title("Prevent Running Hour")
axes3.set_xlabel("running hour")
axes3.set_ylabel("count")

axes4 = fig.add_subplot(2,2,4)
axes4.hist(result_dataset['방지시설연간가동일수'],bins=20)
axes4.set_title("Prevent Running Day")
axes4.set_xlabel("running day")
axes4.set_ylabel("count")
```

```python
plt.plot(result_dataset['좌표정보(x)'], result_dataset['좌표정보(y)'], 'o')
```

## 탐색적 자료분석

```python
result_dataset[['배출시설조업시간', '배출시설연간가동일수','방지시설조업시간', '방지시설연간가동일수']].mean()
```

```python
rank_hour = result_dataset[['종별명', '배출시설조업시간','방지시설조업시간']].groupby(['종별명']).mean()
rank_day = result_dataset[['종별명', '배출시설연간가동일수','방지시설연간가동일수']].groupby(['종별명']).mean()
```

```python
fig = plt.figure(figsize=(12,6))

axes1 = fig.add_subplot(1,2,1)
axes1.plot(rank_hour)
axes1.set_title("Running Hour")
axes1.set_xlabel("rank")
axes1.set_ylabel("mean")

axes2 = fig.add_subplot(1,2,2)
axes2.plot(rank_day)
axes2.set_title("Running Day")
axes2.set_xlabel("rank")
axes2.set_ylabel("mean")
```
